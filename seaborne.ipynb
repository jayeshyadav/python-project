```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('cars.csv')
df
```

```python
df.columns
```

```python
#line plot 
x=np.arange(10)
y=np.sqrt(x)
plt.figure(figsize=(7,7))
sns.lineplot(x,y)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.distplot(df['width'],bins=10)
plt.show()
```

```python
width_mean=df['width'].mean()
width_median=df['width'].median()
plt.figure(figsize=(7,7))
sns.distplot(df['width'])
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='green')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,x='make')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,x='make')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,y='make')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df,x='body-style')
plt.show()
```

```python
#plt.figure(figsize=(7,7))
sns.countplot(data=df,x='make',palette='Spectral_r')
plt.xticks(rotation=75)

plt.show()
```

```python
plt.figure(figsize=(9,9))
sns.countplot(data=df,x='make',hue='body-style')
plt.xticks(rotation=75)

plt.show()
```

```python
df.corr()
```

```python
cor=df.corr()
plt.figure(figsize=(7,7))
sns.heatmap(cor)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x='width',y='price')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x='width',y='price',hue='body-style')
plt.show()
```

```python
plt.figure(figsize=(10,4))
#box plot can be done either one column or two columns
sns.boxplot(data=df, x="price") 
plt.show()
```

```python
plt.figure(figsize=(10,4))
#if you wish to increase the range add whis parameter, by default it is 1.5
sns.boxplot(data=df, x="price", whis=3) 
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="price", y="make") 
plt.show()
```

```python
x = [1,2,3,4]
y = [3,4,5,6]

#subplot accepts number of rows and columns
fig,axes = plt.subplots(2,2,figsize=(8,6))

#axes symbolizes the matrix cordinates
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)

plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,10))
sns.distplot(df['price'],ax=axes[0,0])
sns.countplot(data=df,y='make',ax=axes[0,1])
sns.scatterplot(data=df,y='width',x='price',hue='body-style',ax=axes[1,0])
sns.boxplot(data=df,x='price',ax=axes[1,1])
plt.show()
```
