```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title('area vs price')
plt.xlabel('area')
plt.ylabel('price')
plt.xlim((400,2000))
plt.ylim((10,50))
plt.xticks(rotation=45)
plt.yticks(rotation=45)


plt.show()
```

```python
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)
plt.figure(figsize=(6,6))
plt.plot(x,y1,color='green')
plt.plot(x,y2,color='red')
plt.plot(x,x,color='blue')
plt.legend(labels=['linear','sqrt','log'])
plt.show()
```

```python
x=np.arange(0,20)
y1=np.arange(5,15,0.5)
y2=np.log(x)
plt.figure(figsize=(6,6))
plt.plot(x,y1,color='green')
plt.plot(x,y2,color='red')
plt.legend(labels=['sqrt','log'])
plt.show()
```

```python
#histogram
df=pd.read_csv('cars.csv')
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,7))
plt.title('width frequency distrubution')
plt.xlabel('width')
plt.ylabel('frequency')
plt.hist(df['width'],bins=10)
plt.show()
```

```python
df['width'].hist(bins=10)
plt.show()
```

```python
width_mean=df['width'].mean()
width_median=df['width'].median()
```

```python
plt.figure(figsize=(7,7))
plt.title('width frequency distrubution')
plt.xlabel('width')
plt.ylabel('frequency')
plt.hist(df['width'],bins=10)
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='green')
df['width'].hist(bins=10)

plt.show()
```
