```python
#understanding attributes of matplot
figsize:to give size to chart area 
labels:to assign label to data(xlabel,ylabel)
title:to give title of chart
limits:to start and end point to each axis (xlim,ylim)
ticks:used to rotate the charater visible on x and y axis(xticks,yticks)
legend:give label to plot
color:used to givecolor to plot (function parameter)
```

```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]
#increase size of plot 
#plt.figure(figsize=(width,height))
plt.figure(figsize=(7,7))
plt.plot(area,price)
#add information of the graph 
#add title
plt.title('area vs price')
#add label for y and x 
plt.xlabel('area')
plt.ylabel('price')
#adding limit to the plot
plt.xlim((400,2000))
plt.ylim((5,50))
plt.show()
```

```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]
#increase size of plot 
#plt.figure(figsize=(width,height))
plt.figure(figsize=(7,7))
plt.plot(area,price)
#add information of the graph 
#add title
plt.title('area vs price')
#add label for y and x 
plt.xlabel('area')
plt.ylabel('price')
#adding limit to the plot
plt.xlim((400,2000))
plt.ylim((5,50))
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.show()
```

```python
#multiple plots
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)
plt.figure(figsize=(6,6))
#changing the color of the line 
plt.plot(x,y1,color='yellow')
plt.plot(x,y2,color='red')
#legend in 2 line ploting 
plt.legend(labels=['sqrt','log'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y1,color='yellow')
plt.plot(x,y2,color='red')
plt.plot(x,x,color='blue')
plt.legend(labels=['linear','sqrt','log'])
#legend automatically takes the color just enter a list in which you plot the line 
plt.show()
```

```python
#histograms plot :
#this is used for continuous data .hist is used to define histogram type plot.
#more attribute of matplotlib:
#axvline : give verticle line on x axis
#bins:define number of bins size required in plot (function parameter of hist)
```

```python
df=pd.read_csv('cars.csv')
df.info()
```

```python
df.describe()
```

```python
#before make graph write down your intution using describe function is the dat normally distrubuted
plt.figure(figsize=(7,7))
plt.title('width frequency distrubution')
plt.ylabel('freqeuncy')
plt.xlabel('width')
#bins are no of bars in histogram 
plt.hist(df['width'],bins=10)
plt.show()
```

```python
#pandas is made from numpy and matplotlib so all stuff we preform with matplotlib
df['width'].hist(bins=15)
plt.show()
```

```python
#show mean and median plot in graph
width_mean=df['width'].mean()
width_median=df['width'].median()
```

```python

plt.figure(figsize=(7,7))
plt.hist(df['width'],bins=10)
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='green')
plt.title('width frequency distrubution')
plt.ylabel('freqeuncy')
plt.xlabel('width')
plt.show()
```

```python
#above same ploting using pandas
plt.figure(figsize=(7,7))
df['width'].hist(bins=10)
#plt.hist(df['width'],bins=10)
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='green')
plt.title('width frequency distrubution')
plt.ylabel('freqeuncy')
plt.xlabel('width')
plt.show()
```

```python
#Bar plot
#Majorly used for catogorial data
#kind:type of plot specify bar in kind will give bar plot (bar,barh)
```

```python
#here the kind accepts bar this will create verticals bars
plt.figure(figsize=(9,9))
df['make'].value_counts().plot(kind='bar')
plt.title('make vs count')
plt.xlabel('make')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(9,9))
df['make'].value_counts().plot(kind='barh')
plt.title('make vs count')
plt.xlabel('make')
plt.ylabel('count')
plt.show()
```

```python
#pie chart 
#used to define percentage based data .pie function used to make pie plot 
#Autopct : used to give percentage distrubution over data 
#Explode :used to give extract pie out 
```

```python
#this is to remove a part of sector aways from the pie chart.
#with use of explode attribute it is possible 
#the parameter in explode moves from top anticlockwise using pandas 
plt.figure(figsize=(7,7))
df['body-style'].value_counts().plot(kind='pie')
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['body-style'].value_counts().plot.pie(autopct='%1.1f%%',explode=(0,0,0,0,0.1))
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['body-style'].value_counts().plot(kind='pie',autopct='%1.1f%%',explode=(0,0,0,0,0.1))
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['body-style'].value_counts().plot(kind='pie',autopct='%1.2f%%',explode=(0,0,0,0.2,0.1))
plt.show()
```
