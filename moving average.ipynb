```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('HDFCALLN.csv')
df
```

```python
dg=df[df['Series']=='EQ']
dg
```

```python
new=pd.DataFrame(dg['Close Price'])
new
```

```python
day= pd.Series(range(1,641),dtype='int32')
pd.DataFrame(day, columns=['Name'])
```

```python
new['days']=day
new
```

```python
new= new[['days','Close Price']]
new
```

```python
new
```

```python
for i in range(0,new.shape[0]-2):
    new.loc[new.index[i+2],'SMA_10'] = np.round(((new.iloc[i,1]+ new.iloc[i+1,1] +new.iloc[i+2,1])/10),1)
```

```python
new
```

```python
new['pandas_SMA_10'] = new.iloc[:,1].rolling(window=10).mean()
new
```

```python
for i in range(0,new.shape[0]-3):
    new.loc[new.index[i+3],'SMA_20'] = np.round(((new.iloc[i,1]+ new.iloc[i+1,1] +new.iloc[i+2,1]+new.iloc[i+3,1])/20),1)
new
```

```python
new['pandas_SMA_20'] = new.iloc[:,1].rolling(window=20).mean()
new
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=[25,20])
plt.grid(True)
plt.plot(new['Close Price'],label='data')
plt.plot(new['pandas_SMA_10'],label='SMA 10 days')
plt.plot(new['pandas_SMA_20'],label='SMA 20 days')
plt.legend(loc=2)
```

```python
plt.figure(figsize=[15,10])
plt.grid(True)
plt.plot(new['Close Price'],label='HDFC Bank')
plt.plot(new['pandas_SMA_10'],label='SMA 10 days')
plt.plot(new['pandas_SMA_20'],label='SMA 20 days')
plt.legend(loc=2)
```
